## Testing of Dark Matter Redshift Space Power Spectrum 

**Date**: 3/21/14 

**Goal**: Reproduce the figures from Vlah et al. 2012 for the dark matter redshift space power spectrum at multiple redshifts

```python
# import the necessary modules
from pyPT.rsd import power_dm
from pyPT.cosmology import power
```

```python
# initialize the power spectrum with correct cosmo params
cosmo = {'force_flat' : True, 'default' : 'WMAP5', 'h' : 0.701, 'n' : 0.96, 'sigma_8' : 0.807, 'omegac': (1-0.165)*0.279, 'omegal': 0.721, 'omegab': 0.165*0.279, 'omegar': 0.}
P = power_dm.DMSpectrum(include_2loop=True, num_threads=4, transfer_fit="CAMB", cosmo=cosmo)
```

```python
P.update(max_mu=8)
```

```python
# set up the velocities at z = 0
P.sigma_bv2 = 375.
P.sigma_bv4 = 432.
P.sigma_v2 = 209.
```

```python
# initialize the no-wiggle linear power spectrum
Plin = power.Power(k=P.k, z=0., cosmo=P.cosmo, transfer_fit="EH_no_wiggles")
P_nw = Plin.power
```

```python
# now plot P00, and compare to Fig. 1 of Vlah et al. 2012

#----------------------------------------------------------------------------------
# DO Z = 0 ON THE FIRST SUBPLOT
#----------------------------------------------------------------------------------
ax1 = subplot(211)
P.update(z=0)
norm = P.D**2 * P_nw
plot(P.k, P.P00.total.mu0 / norm)
plot(P.k, (P.D**2 * P.power.power) / norm, c='k', ls='--')

#----------------------------------------------------------------------------------
# DO Z = 1 ON THE SECOND SUBPLOT
#----------------------------------------------------------------------------------
ax2 = subplot(212)

P.update(z=1.0)
norm = P.D**2 * P_nw
plot(P.k, P.P00.total.mu0 / norm)
plot(P.k, (P.D**2 * P.power.power) / norm, c='k', ls='--')

# make it look nice
ax1.set_xlim(0., 0.3)
ax1.set_ylim(0.8, 1.5)
ax1.minorticks_on()
ax1.tick_params(which='major', width=2, length=8)
ax1.tick_params(which='minor', width=1, length=4)

ax2.set_xlim(0., 0.3)
ax2.set_ylim(0.8, 1.5)
ax2.minorticks_on()
ax2.tick_params(which='major', width=2, length=8)
ax2.tick_params(which='minor', width=1, length=4)

ax2.set_xlabel(r"$\mathrm{k \ (h/Mpc)}$", fontsize=16)
ax1.set_ylabel(r"$P_{00}^{ss} \ / P_\mathrm{NW}$", fontsize=16)
ax2.set_ylabel(r"$P_{00}^{ss} \ / P_\mathrm{NW}$", fontsize=16)
ax1.set_title(r"$\mathrm{Comparison \ of \ P^{ss}_{00}}$", fontsize=20) 

figtext(0.2, 0.85, r"$z = 0$", fontsize=20)
figtext(0.2, 0.4, r"$z = 1.0$", fontsize=20)
```

```python
# now plot P01 and compare to Fig. 2
ax1 = subplot(211)
P.update(z=0.)

norm = 2*P.f*P.D**2 * P_nw
plot(P.k, P.P01.total.mu2/norm)
plot(P.k, 2*P.f*P.D**2*P.power.power/norm, c='k', ls='--')

ax2 = subplot(212)
P.update(z=1.)

norm = 2*P.f*P.D**2 * P_nw
plot(P.k, P.P01.total.mu2/norm)
plot(P.k, 2*P.f*P.D**2*P.power.power/norm, c='k', ls='--')

# make it look nice
ax1.set_xlim(0, 0.3)
ax1.set_ylim(0.8, 1.5)
ax1.minorticks_on()
ax1.tick_params(which='major', width=2, length=8)
ax1.tick_params(which='minor', width=1, length=4)

ax2.set_xlim(0, 0.3)
ax2.set_ylim(0.8, 1.5)
ax2.minorticks_on()
ax2.tick_params(which='major', width=2, length=8)
ax2.tick_params(which='minor', width=1, length=4)

ax2.set_xlabel(r"$\mathrm{k \ (h/Mpc)}$", fontsize=16)
ax2.set_ylabel(r"$P^{ss}_{01}[\mu^2] \ / \ (2 \ f \ P_\mathrm{NW})$", fontsize=16)
ax1.set_ylabel(r"$P^{ss}_{01}[\mu^2] \ / \ (2 \ f \ P_\mathrm{NW})$", fontsize=16)
ax1.set_title(r"$\mathrm{Comparison \ of \ P^{ss}_{01}}$", fontsize=20) 

figtext(0.2, 0.85, r"$z = 0$", fontsize=20)
figtext(0.2, 0.4, r"$z = 1.0$", fontsize=20)
```
