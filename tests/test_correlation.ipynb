## Testing of Correlation Function

**Date Started**: 5/4/14

**Goal**: Test the Fourier transform of power spectrum to produce correlation function multipoles $\xi_0$, $\xi_2$, as well as 2D $\xi(k, \mu)$

**Notes**:

* *5/7*: The correlation function multipoles appear to be working well for halos.

```python
# import the necessary modules
from pyRSD.rsd import power_halo, correlation
from pyRSD.cosmology import power
from pyRSD.data import *
```

```python
def make_pretty(ax, legend_size=16, legend_loc=0):
    if legend_loc is not None:
        ax.legend(loc=legend_loc, prop={'size':legend_size})
    ax.minorticks_on()
    ax.tick_params(which='major', width=2, length=8)
    ax.tick_params(which='minor', width=1, length=4)
```

```python
# read in the simulation data, at z = 0
P00_mu0_data_z0 = P00_mu0_z_0_000()
P01_mu2_data_z0 = P01_mu2_z_0_000()
P11_mu4_data_z0 = P11_mu4_z_0_000()
Pdv_mu0_data_z0 = Pdv_mu0_z_0_000()
```

```python
# initialize the power spectrum with correct cosmo params
cosmo = {'force_flat' : True, 'default' : 'WMAP5', 'h' : 0.701, 'n' : 0.96, 'sigma_8' : 0.807, 'omegac': (1-0.165)*0.279, 'omegal': 0.721, 'omegab': 0.165*0.279, 'omegar': 0.}
P = power_halo.HaloSpectrum(k=logspace(-3, np.log10(0.5), 500), num_threads=6, transfer_fit="CAMB", cosmo=cosmo, max_mu=6)
```

```python
# initialize the sigmas
P.sigma_bv2 = P.sigma_bv4 = P.sigma_v2 = 0.
```

```python
# load in the data
def load_z0_sims():
    P.load(P00_mu0_data_z0[:,0], P00_mu0_data_z0[:,1], 'P00', 'mu0')
    P.load(P01_mu2_data_z0[:,0], P01_mu2_data_z0[:,1], 'P01', 'mu2')
    P.load(P11_mu4_data_z0[:,0], P11_mu4_data_z0[:,1], 'P11', 'mu4')
    P.load(Pdv_mu0_data_z0[:,0], Pdv_mu0_data_z0[:,1], 'Pdv', None)
```

```python
# initialize the no-wiggle linear power spectrum
Plin = power.Power(k=P.k, z=0., cosmo=P.cosmo, transfer_fit="EH_no_wiggles")
P_nw = Plin.power
```

```python
# set up the bias related variables
biases = [1.18, 3.05]
P.b1 = biases[0]

load_z0_sims()
```

```python
# initialize the correlation function class
corr = correlation.Correlation(P)

# the range of s we want
s = logspace(log10(0.5), log10(150), 200)
```

```python
# compute the monopoles for all biases
colors = ['DodgerBlue', 'Crimson']
mono_smoothing = [0.5, 0.75]
for i, b in enumerate(biases):
    
    # update the bias
    corr.b1 = b
    
    # the linear monopole
    xi0_linear = corr.monopole(s, smoothing_radius=0., kcut=None, linear=True)
    
    # the full monopole
    xi0 = corr.monopole(s, smoothing_radius=mono_smoothing[i], kcut=0.2, linear=False)
    
    # plot both monopoles
    plot(s, s**2*xi0/corr.b1, label=r'$\mathrm{full, \ b_1 = \ %.2f}$' %b, c=colors[i])
    plot(s, s**2*xi0_linear/corr.b1, ls='--', c=colors[i], label=r'$\mathrm{linear, \ b_1 = \ %.2f}$' %b)

xlim(2, 120)
ylim(0, 140)
make_pretty(gca(), legend_loc='upper right')

xlabel(r"$\mathrm{s \ [Mpc/h]}$", fontsize=18)
ylabel(r"$s^2 \xi_0(s)/b_1 \ \mathrm{[Mpc/h]^2}$", fontsize=18)
title(r"$\mathrm{Correlation \ function \ monopole \ \xi_0(s) \ at \ z=0}$", fontsize=20) 
```

```python
# compute the quadrupole for all biases
offsets = [0., 45.]
quad_smoothing = [0.61, 0.52]

for i, b in enumerate(biases):
    
    # update the bias
    corr.b1 = b
    
    # the linear quadrupole
    xi2_linear = corr.quadrupole(s, smoothing_radius=0., kcut=None, linear=True)
    
    # the full monopole
    xi2 = corr.quadrupole(s, smoothing_radius=quad_smoothing[i], kcut=0.23, linear=False)
    
    # plot both monopoles
    plot(s, -s**2*xi2/corr.b1 + offsets[i], label=r'$\mathrm{full, \ b_1 = \ %.2f}$' %b, c=colors[i])
    plot(s, -s**2*xi2_linear/corr.b1 + offsets[i], ls='--', c=colors[i], label=r'$\mathrm{linear, \ b_1 = \ %.2f}$' %b)

xlim(2, 120)
#ylim(0, 140)
make_pretty(gca(), legend_loc='lower right')

xlabel(r"$\mathrm{s \ (Mpc/h)}$", fontsize=18)
ylabel(r"$-s^2 \xi_2(s)/b_1 \ \mathrm{[Mpc/h]^2}$", fontsize=18)
title(r"$\mathrm{Correlation \ function \ quadrupole \ \xi_2(s) \ at \ z=0}$", fontsize=20)
```
