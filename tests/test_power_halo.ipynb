## Testing of Halo Redshift Space Power Spectrum 

**Date Started**: 3/23/14

**Goal**: Reproduce the figures from Vlah et al. 2013 for the halo redshift space power spectrum at multiple redshifts

**Notes**:

* *5/1/14*: Working version of the halo-halo RSD code, with the following features:
    
    * stochasticity and higher order biases determined as function of linear bias via interpolation
    * terms up to $\mu^6$ included and figures reproduced in Vlah et al. 2013
    * simulation data can now be imported from a pyRSD.data module

```python
# import the necessary modules
from pyRSD.rsd import power_halo
from pyRSD.cosmology import power
from pyRSD.data import *
```

```python
def make_pretty(ax, legend_size=16, legend_loc=0):
    ax.legend(loc=legend_loc, prop={'size':legend_size})
    ax.minorticks_on()
    ax.tick_params(which='major', width=2, length=8)
    ax.tick_params(which='minor', width=1, length=4)
```

```python
# read in the simulation data, at z = 0
P00_mu0_data_z0 = P00_mu0_z_0_000()
P01_mu2_data_z0 = P01_mu2_z_0_000()
P11_mu4_data_z0 = P11_mu4_z_0_000()
Pdv_mu0_data_z0 = Pdv_mu0_z_0_000()

# read in the simulation data, at z = 1
P00_mu0_data_z05 = P00_mu0_z_0_509()
P01_mu2_data_z05 = P01_mu2_z_0_509()
P11_mu4_data_z05 = P11_mu4_z_0_509()
Pdv_mu0_data_z05 = Pdv_mu0_z_0_509()
```

```python
# initialize the power spectrum with correct cosmo params
cosmo = {'force_flat' : True, 'default' : 'WMAP5', 'h' : 0.701, 'n' : 0.96, 'sigma_8' : 0.807, 'omegac': (1-0.165)*0.279, 'omegal': 0.721, 'omegab': 0.165*0.279, 'omegar': 0.}
P = power_halo.HaloSpectrum(num_threads=6, transfer_fit="CAMB", cosmo=cosmo, max_mu=6)
P_pt = power_halo.HaloSpectrum(num_threads=6, transfer_fit="CAMB", cosmo=cosmo, max_mu=6)
```

```python
# initialize the sigmas and biases
P.sigma_bv2 = P.sigma_bv4 = P.sigma_v2 = 0.
P_pt.sigma_bv2 = P_pt.sigma_bv4 = P_pt.sigma_v2 = 0

P.bs = P.b1 = P.b2_00 = P.b2_01 = 0.
P_pt.bs = P_pt.b1 = P_pt.b2_00 = P_pt.b2_01 = 0.
```

```python
# load in the data
def load_z0_sims():
    P.load(P00_mu0_data_z0[:,0], P00_mu0_data_z0[:,1], 'P00', 'mu0')
    P.load(P01_mu2_data_z0[:,0], P01_mu2_data_z0[:,1], 'P01', 'mu2')
    P.load(P11_mu4_data_z0[:,0], P11_mu4_data_z0[:,1], 'P11', 'mu4')
    P.load(Pdv_mu0_data_z0[:,0], Pdv_mu0_data_z0[:,1], 'Pdv', None)
    
def load_z05_sims():
    P.load(P00_mu0_data_z05[:,0], P00_mu0_data_z05[:,1], 'P00', 'mu0')
    P.load(P01_mu2_data_z05[:,0], P01_mu2_data_z05[:,1], 'P01', 'mu2')
    P.load(P11_mu4_data_z05[:,0], P11_mu4_data_z05[:,1], 'P11', 'mu4')
    P.load(Pdv_mu0_data_z05[:,0], Pdv_mu0_data_z05[:,1], 'Pdv', None)
```

```python
# initialize the no-wiggle linear power spectrum
Plin = power.Power(k=P.k, z=0., cosmo=P.cosmo, transfer_fit="EH_no_wiggles")
P_nw = Plin.power
```

```python
# now plot P00_hh, and compare to Fig. 4 of Vlah et al. 2013

#----------------------------------------------------------------------------------
# DO Z = 0 ON THE FIRST SUBPLOT
#----------------------------------------------------------------------------------
ax1 = subplot(211)
load_z0_sims()
P.update(z=0, b1=3.05)
P_pt.update(z=0, b1=3.05)

axhline(y=1, c='k', ls='--')
norm = (P.b1**2*P.P00.total.mu0)
plot(P.k, (P.P00_ss.total.mu0) / norm , label='with sim data')
plot(P_pt.k, (P_pt.P00_ss.total.mu0) / norm, label='pure PT')

#----------------------------------------------------------------------------------
# DO Z = 1 ON THE SECOND SUBPLOT
#----------------------------------------------------------------------------------
ax2 = subplot(212)
load_z1_sims()
P.update(z=0.987, b1=4.64)
P_pt.update(z=0.987, b1=4.64)

axhline(y=1, c='k', ls='--')
norm = (P.b1**2*P.P00.total.mu0)
plot(P.k, (P.P00_ss.total.mu0) / norm , label='with sim data')
plot(P_pt.k, (P_pt.P00_ss.total.mu0) / norm, label='pure PT')

#----------------------------------------------------------------------------------
# MAKE IT LOOK NICE
#----------------------------------------------------------------------------------
ax1.set_xlim(0., 0.3)
ax1.set_ylim(0.5, 1.1)
make_pretty(ax1, legend_loc='lower left')

ax2.set_xlim(0., 0.3)
ax2.set_ylim(0.6, 1.2)
make_pretty(ax2, legend_loc='lower left')

ax2.set_xlabel(r"$\mathrm{k \ (h/Mpc)}$", fontsize=16)
ax2.set_ylabel(r"$(P_{00}^\mathrm{hh} - \bar{n}^{-1}) \ / \ (b_1^2 P_{00})$", fontsize=16)
ax1.set_ylabel(r"$(P_{00}^\mathrm{hh} - \bar{n}^{-1}) \ / \ (b_1^2 P_{00})$", fontsize=16)
ax1.set_title(r"$\mathrm{Comparison \ of \ P_{00}^\mathrm{ss, hh}[\mu^0]}$", fontsize=20) 

figtext(0.8, 0.8, r"$z = 0$", fontsize=20)
figtext(0.8, 0.25, r"$z = 1.0$", fontsize=20)
```
